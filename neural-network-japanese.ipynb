```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

def load_dataset():
    with h5py.File("./train_test_file_list.h5") as f:
        train_x = np.array(f.get('train_x'))
        train_y = np.array(f.get('train_y'))
        test_x = np.array(f.get('test_x'))
        test_y = np.array(f.get('test_y'))
    return train_x, train_y, test_x, test_y

train_x, train_y, test_x, test_y = load_dataset()
train_x.reshape(train_x.shape[0], 1, 28, 28)
```
